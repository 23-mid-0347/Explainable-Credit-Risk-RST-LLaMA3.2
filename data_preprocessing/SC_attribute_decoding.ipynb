```python
import pandas as pd

df = pd.read_csv("D:\\Coding\\6th sem\\Soft Computing\\Credit Risk Assessment\\data\\Cleaned_german_data.csv")
```

```python
#Checking Acc Status
df["CheckingAccountStatus"] = df["CheckingAccountStatus"].map({
    1: "< 0 DM",
    2: "0 <= balance < 200 DM",
    3: ">= 200 DM",
    4: "No checking account"
})
```

```python
print(df["CreditHistory"].unique())
```

```python
#Credit History
df["CreditHistory"] = df["CreditHistory"].map({
    0: "No credit history",
    1: "All credits paid back",
    2: "Existing credits paid back",
    3: "Delay in paying",
    4: "Critical account"
})
print(df["CreditHistory"].unique())

```

```python
#Purpose
df["Purpose"] = df["Purpose"].map({
    0: "Car (new)",
    1: "Car (used)",
    2: "Furniture/equipment",
    3: "Radio/TV",
    4: "Domestic appliances",
    5: "Repairs",
    6: "Education",
    7: "Vacation",
    8: "Retraining",
    9: "Business",
    10: "Others"
})
```

```python
#savings ACC
df["SavingsAccount"] = df["SavingsAccount"].map({
    1: "< 100 DM",
    2: "100 <= savings < 500",
    3: "500 <= savings < 1000",
    4: ">= 1000 DM",
    5: "Unknown / None"
})
```

```python
#employment Details
df["EmploymentDuration"] = df["EmploymentDuration"].map({
    1: "Unemployed",
    2: "< 1 year",
    3: "1 - 4 years",
    4: "4 - 7 years",
    5: ">= 7 years"
})
```

```python
#Personal Status
df["PersonalStatus"] = df["PersonalStatus"].map({
    1: "Male divorced/separated",
    2: "Female divorced/separated/married",
    3: "Male single",
    4: "Male married/widowed"
})
```

```python
#Guarantor
df["Guarantor"] = df["Guarantor"].map({
    1: "None",
    2: "Co-applicant",
    3: "Guarantor"
})
```

```python
#Property
df["Property"] = df["Property"].map({
    1: "Real estate",
    2: "Building society savings",
    3: "Car or other",
    4: "Unknown / None"
})
```

```python
#Other installment Plans
df["OtherInstallmentPlans"] = df["OtherInstallmentPlans"].map({
    1: "Bank",
    2: "Stores",
    3: "None"
})
```

```python
#Housing
df["Housing"] = df["Housing"].map({
    1: "Rent",
    2: "Own",
    3: "Free"
})
```

```python
#Job
df["Job"] = df["Job"].map({
    1: "Unemployed",
    2: "Unskilled",
    3: "Skilled",
    4: "Highly skilled"
})
```

```python
#telephone
df["Telephone"] = df["Telephone"].map({
    1: "No",
    2: "Yes"
})
```

```python
#Foreign worker
df["ForeignWorker"] = df["ForeignWorker"].map({
    1: "Yes",
    2: "No"
})
```
