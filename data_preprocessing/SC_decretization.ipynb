```python
import pandas as pd

df = pd.read_csv("D:\\Coding\\6th sem\\Soft Computing\\Credit Risk Assessment\\data\\Decoded_german_data.csv", keep_default_na=False)
print(df["Guarantor"].unique())
print(df["OtherInstallmentPlans"].unique())
```

```python
print(df.isnull().sum())
```

```python
#AGE
df["AgeCategory"] = pd.qcut(
    df["Age"],
    q=3,
    labels=["Young", "Middle-aged", "Senior"],
    duplicates="drop"
)
```

```python
#Load Duration
df["DurationCategory"] = pd.qcut(
    df["Duration"],
    q=3,
    labels=["Short", "Medium", "Long"],
    duplicates="drop"
)
```

```python
#Credit Amount
df["CreditAmountCategory"] = pd.qcut(
    df["CreditAmount"],
    q=3,
    labels=["Low", "Medium", "High"],
    duplicates="drop"
)
```

```python
print(df[["AgeCategory",
          "DurationCategory",
          "CreditAmountCategory"]].isnull().sum())
```

```python
#Before RST
df_rst = df.drop(columns=["Age", "Duration", "CreditAmount"])
```

```python
print(df_rst.isnull().sum())

```
